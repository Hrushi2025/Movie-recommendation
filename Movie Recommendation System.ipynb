```python
# Required labs 
import numpy as np 
import pandas as pd 
```

```python
# our data set to be used 
Movies = pd.read_csv('tmdb_5000_movies.csv')
Credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# we merged the Movies and Credit data on the title and made on df named as movies
movies= Movies.merge(Credits,on = 'title')
```

```python
# need to select columns which are required further to make recommendation 
# geners
# id
# keywords
# title
# overview
# cast 
# crew
```

```python
# here we had made the columns which are required further 
movies =movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# we will any null vales are present in data
movies.isnull().sum() # and we get 3 null at overview we can drop it not that big impact on data as it is only 3 
```

```python
# we have droped those rows 
movies.dropna(inplace=True)
```

```python
# by using apply function and ast lab of in function we are able to see this word as excpected 
movies['genres']= movies['genres'].apply(convertor) 
```

```python
# same we did with keywords column
movies['keywords'] = movies['keywords'].apply(convertor) 
```

```python
movies.head()
```

```python
# let do for now cast column we need first 3 actor which are important one for us, now we will make one more function
import ast

def convertor3(obj):
    L = []
    for i in ast.literal_eval(obj)[:3]:  # Take only first 3 elements
        L.append(i['name'])
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertor3)
```

```python
movies.head()
```

```python
# we have to for the crew column but where as it is little diffcult coz i dont want all the dict i want only 
# those dict where the job: director and need to extract specific name of it so we will make on function for it 
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# now we will make list of overview column strings 
movies['overview'] =movies['overview'].apply(lambda x:x.split()) # now finally is is in list
```

```python
movies.head()
```

```python
#now need remove white spaces from columns name genres.keywords.cast.crew we are doing this to avoid output of same name 
# for example searching for sam which and getting sam mendis we have to avoide this so we will remove white spaces 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head() # so finally we are done with the white spaces 
```

```python
# now we will create one more column named as tags
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
# now we will take required data 
new_df =movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#Now we have make lower case of each letter for recommendation 
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer #we are using this lab for removing stem ending of same meaning words
ps = PorterStemmer()
```

```python
# this function we have made for the remove the stem ending 
def stem(text):
    y= []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem) #now we have applied this to the new_df
```

```python
# max freature for how many word we want to work 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000,stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()# converted in numpy array and maked vectors 
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
ps.stem('writing') # stem is removed of every word 
```

```python
# we will use cosine similarity between two vectors 
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity =cosine_similarity(vectors) # we have got the shape of it same as movie 
```

```python
#we cant sort directly coz we are doing all this on the bases of index and so we will use enumerates function 
# now we will create a tuple where 1st movie could compare with each other easily and use the lambda function to check 1 to all movies
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#Now we will make one fun to recommend the movies based on its index with there distance 
# like for example if anyone provides me a movie and i have more 5 suggestion related to same movie 

def recommd(movie):
      movie_index = new_df[new_df['title']== movie].index[0]
      distance =similarity[movie_index]
      movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
      for i in movies_list:
          print(new_df.iloc[i[0]].title)# due to this we can get titles too and without it only using[i[0]] we get tag
           
      
```

```python
recommd('Avatar') # here we have checked that we can get the top 5 recommendation same as this movie we are getting their tags
```

```python
recommd('Avatar') # now you can see we are getting title too
```
